# Scatterplots with Seaborn

In the cell below, we will learn how to use scatterplots with Seaborn. Scatterplots are a useful visualization tool for exploring the relationship between two continuous variables. They can help us identify patterns, trends, and outliers in the data.

With Seaborn, we can create scatterplots using the `scatterplot()` function. This function allows us to plot the relationship between two variables on a Cartesian plane, where each data point is represented by a dot.

In this example, we will be using the famous Iris dataset, which contains measurements of sepal length, sepal width, petal length, and petal width for three different species of iris flowers.

To create a scatterplot, we need to specify the variables we want to plot on the x-axis and y-axis. In our case, we will plot the sepal length on the x-axis and the sepal width on the y-axis.

Additionally, we will explore the `pairplot()` function, which allows us to create a grid of scatterplots for multiple combinations of variables. This can be helpful when we want to visualize the relationships between multiple variables simultaneously.

Let's dive into the code and create some scatterplots using Seaborn!


```python
# Load iris data from Seaborn
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.info()
```

```python
# find unique species
iris['species'].unique()
```

```python
# scatterplot of sepal length and width
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = iris)
```

```python
# The data above is not very useful, so we can color the points by species
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = iris, hue = 'species')
plt.show()
```

```python
# We can also use the pairplot function to plot all the scatterplots for the different combinations of variables
sns.pairplot(iris, hue = 'species')
plt.show()
```

```python
# chnage the color palette
sns.pairplot(iris, hue = 'species', palette = 'husl')
```

```python
# get the flights data from seaborn
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.info()
```

```python
may_flights = flights.query("month == 'May'")
may_flights
```

```python
# line plot of the number of passengers over time
plt.figure(figsize = (10, 6))
sns.lineplot(x = 'year', y = 'passengers', data = may_flights)
plt.show()
```

```python
# plot all the months
plt.figure(figsize = (10, 6))
sns.lineplot(x = 'year', y = 'passengers', data = flights, hue = 'month')
plt.show()
```

```python
# pass the flight data without hue
# This will plot the mean number of passengers for each year and the 95% confidence interval
plt.figure(figsize = (10, 6))
sns.lineplot(x = 'year', y = 'passengers', data = flights)
```

```python
# sine wave and cosine wave
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# plot the sine wave and cosine wave
plt.figure(figsize = (10, 6))
sns.lineplot(y1)
sns.lineplot(y2)
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```
